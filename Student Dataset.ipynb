```python
data=pd.read_csv("C:\AKS study original\Student_performance_data _.csv")
data
```

```python
data.info()
```

```python
data['Age'].mean()
plt.hist(data['Age'])
plt.title('student ages ',loc='right')
plt.xlabel('Age')
plt.ylabel('no. of students')
```

```python
b=[[],[],[],[]]
count=0
age_mean=[]
for i in range(15,19):
    for j in data['Age']:
        if j==i:
            b[i-15].append(data['GPA'][count])
        count+=1
    count=0
for i in range(4):
    b[i]=pd.Series(b[i])
    age_mean.append(b[i].mean())
plt.plot([15,16,17,18],age_mean,'o:b')
plt.xlabel('age')
plt.ylabel('GPA')
plt.title('Age v/s GPA')
```

```python
#data shows no. of female students are slightly more than males
plt.hist(data['Gender'])
plt.xlabel('Gender:0 is male 1 is female')
plt.ylabel('No. of students')
```

```python
b=[[],[]]
count=0
gender_mean=[]
for i in range(2):
    for j in data['Gender']:
        if j==i:
            b[i].append(data['GPA'][count])
        count+=1
    count=0
for i in range(2):
    b[i]=pd.Series(b[i])
    gender_mean.append(b[i].mean())
plt.bar([0,1],gender_mean,width=0.1)
plt.xlabel('Gender:0 is male 1 is female')
plt.ylabel('GPA')
plt.title('genderwise GPA')
gender_mean
#the mean GPA is almost same for both the genders
```

```python
#most students are caucasian 
data['Ethnicity'].plot(kind='hist')
plt.title('student ethnicity',loc='right')
plt.xlabel('0:Caucasian 1:AfricanAmerican 2:Asian 3:Other')
plt.ylabel('no. of students')
```

```python

plt.scatter(data['StudyTimeWeekly'],data['GPA'])
```

```python
#from graph it is apparent that in general students with higher grades have less absences
plt.scatter(data['Absences'],data['GPA'],c='hotpink')
plt.xlabel('absences')
plt.ylabel('GPA')
plt.title('absences vs GPA')
```

```python
#majority of students don't have tutoring
data['Tutoring'].plot(kind='hist')
plt.ylabel('no. of students')
plt.xlabel('0:no 1:yes')
plt.title('TUTORING')
```

```python
list=[[],[]]
count=0
for i in range(2):
    for j in data['Tutoring']:
        if i==j:
            list[i].append(data['GPA'][count])
        count+=1
    count=0
def f(x):
    mean=[]
    for i in range(len(x)):
        x[i]=pd.Series(x[i])
        mean.append(x[i].mean())
    return mean
f(list)
plt.bar(['No Tutoring','Tutoring'],f(list),width=0.7,color='yellow')# We can see students with tutoring have higher avg GPA
plt.xlabel('Tutoring')# hence we can draw a conclusion that tutoring boosts grades
plt.ylabel('GPA')
plt.title('Comparison of average GPA of students with and without tutoring')
```

```python
b=[[],[],[],[],[]]
count=0
for i in range(5):
    for j in data['ParentalSupport']:
        if j==i:
            b[i].append(data['GPA'][count])
        count+=1
    count=0

mean_list=[]
for i in range(5):
    b[i]=pd.Series(b[i])
    mean_list.append(b[i].mean())
mean_list
plt.plot([0,1,2,3,4],mean_list,'o--y')#from this graph we can see students with higher parental supprt are scoring higher gpa on average
plt.ylabel('GPA')
plt.xlabel('0:none 1:low 2:moderate 3:high 4:very high')
plt.title('Graph of parental care v/s the mean GPA of students')
```
